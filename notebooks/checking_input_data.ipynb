```python
mismatch_df = df[df["Target"] != df["Target.1"]]

# Drop NaN comparisons if needed
# mismatch_df = df[~(df["Target"] == df["Target.1"]) & ~(df["Target"].isna() & df["Target.1"].isna())]

# Print names and mismatching targets
print("\nüîç Rows where 'Target' and 'Target.1' differ:")
print(mismatch_df[["Name", "Furnace", "Furnace.1"]].to_string(index=False))

# Print count
print(f"\nTotal mismatches: {len(mismatch_df)}")
```

```python
columns_to_update = ['Target', 'Precursors', 'Temperature (C)', 'Dwell Duration (h)']

for col in columns_to_update:
    col_1 = f"{col}.1"
    if col_1 in df.columns:
        df[col] = df[col_1]  # Replace values in base column with those from `.1` version

# Optionally drop the `.1` columns afterward
extra_columns_to_drop = [f"{col}.1" for col in columns_to_update] + ["Run Name in ALAB.1", "Furnace.1"]
df.drop(columns=extra_columns_to_drop, inplace=True)

# (Optional) Save back to CSV
df.to_csv("../data/alab_synthesis_data/synthesis_TRI.csv", index=False)
```
