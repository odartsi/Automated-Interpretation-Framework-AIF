```python
import pandas as pd
from collections import defaultdict



# Step 1: Load both CSVs
evaluation_df = pd.read_csv("../data/alab_synthesis_data/Dara-AIF_evaluation.csv")
log_df = pd.read_csv("../data/alab_synthesis_data/AIFvsDara_LOG_updated.csv")

# Step 2: Merge on "Run Name in ALAB"
merged_df = pd.merge(
    evaluation_df,
    log_df[['Run Name in ALAB', 'Interpr A - explanation', 'Interpr B - explanation']],
    on='Run Name in ALAB',
    how='left'
)

# Step 3: Define function to classify preferred interpretation
def determine_preferred_source(row):
    pref = str(row['Prefered Interpretation']).strip().lower()
    a_source = str(row['Interpr A - explanation']).strip().lower()
    b_source = str(row['Interpr B - explanation']).strip().lower()

    if "both" in pref:
        return "both"
    elif "neither" in pref:
        return "neither"
    elif pref == "a":
        return a_source
    elif pref == "b":
        return b_source
    else:
        return "unknown"

merged_df['Preferred Source'] = merged_df.apply(determine_preferred_source, axis=1)

# Step 5: Group by Preferred Source, Target, and synthesis conditions
group_cols = ['Preferred Source', 'Target', 'Precursor 1', 'Precursor 2', 'Precursor 3',
              'Precursor 4', 'Furnace', 'Temperature (C)', 'Dwell Duration (h)']

grouped = merged_df.groupby(group_cols, dropna=False)

# Step 6: Structure the data for display
tables_by_source = defaultdict(lambda: defaultdict(list))
for keys, group_df in grouped:
    source = keys[0]
    target = keys[1]
    synth_keys = keys[2:]
    tables_by_source[source][target].append((synth_keys, group_df.reset_index(drop=True)))

# Step 7: Display it
for source_group, target_dict in tables_by_source.items():
    print(f"\n========== Preferred Source: {source_group.upper()} ==========")

    for target_index, (target, groups) in enumerate(target_dict.items(), 1):
        print(f"\n--- Target {target_index}: {target} ---")

        for i, (synth_keys, df) in enumerate(groups, 1):
            run_names = df['Run Name in ALAB'].dropna().tolist()
            run_list_str = ", ".join(run_names)

            # Clean precursors
            precursors = [p for p in synth_keys[:4] if pd.notna(p)]
            precursors_str = ", ".join(precursors)
            furnace = synth_keys[4]
            temp = synth_keys[5]
            dwell = synth_keys[6]

            print(f"  Group {i}: {run_list_str}")
            print(f"    Precursors: {precursors_str} | Furnace: {furnace} | Temp: {temp}Â°C | Dwell: {dwell}h")
```

```python
from collections import defaultdict

# STEP 1: Always include all 'neither'
reeval_samples = merged_df[merged_df['Preferred Source'] == 'neither'].copy()

# STEP 2: Build a diverse selection from the rest
diverse_samples = []
seen_groups = set()
# Normalize Preferred Source values: lowercase + strip whitespace
# Step 1: Normalize the preferred source values
merged_df['Preferred Source'] = merged_df['Preferred Source'].str.lower().str.strip()
merged_df['Preferred Source'] = merged_df['Preferred Source'].replace({
    'aif interpretation': 'aif',
    'dara interpretation': 'dara'
})

# Step 2: Define simplified and robust group columns
group_cols = ['Target', 'Precursor 1', 'Precursor 2', 'Furnace', 'Temperature (C)', 'Dwell Duration (h)']  # simplified for debugging

diverse_samples = []
seen_groups = set()

# Step 3: Loop through each interpretation source
for source in ['aif', 'dara', 'both']:
    subset = merged_df[merged_df['Preferred Source'] == source].copy()
    print(f"\nSOURCE: {source} | Subset size: {subset.shape[0]}")

    if subset.empty:
        continue

    # Fill missing values so groupby won't break
    subset[group_cols] = subset[group_cols].fillna("MISSING")

    grouped = subset.groupby(group_cols)

    for keys, group_df in grouped:
        print(f"Group keys: {keys} | Samples: {group_df['Run Name in ALAB'].tolist()}")

        target = group_df['Target'].iloc[0]
        synth_key = tuple(group_df[group_cols].iloc[0].values)
        group_id = (source, target, synth_key)

        if group_id not in seen_groups:
            seen_groups.add(group_id)
            diverse_samples.append(group_df.iloc[0])

# STEP 3: Combine both
reeval_df = pd.concat([reeval_samples] + [pd.DataFrame(diverse_samples)], ignore_index=True)

# Optional: Sort and view
reeval_df = reeval_df.sort_values(by=['Preferred Source', 'Target'])
print(reeval_df[['Run Name in ALAB', 'Target', 'Preferred Source']])
```

```python
from pymatgen.core import Composition

def remove_elements_from_composition(comp, elements_to_remove):
    composition = Composition(comp)
    filtered = {el: amt for el, amt in composition.get_el_amt_dict().items() if el not in elements_to_remove}
    return Composition(filtered)

def normalize_composition(comp):
    return comp.fractional_composition

def calculate_composition_balance_score_refined(target_comp, output_comp):
    """
    Compare two compositions and compute a similarity score based on element-wise differences.
    """
    elements = set(target_comp.elements).union(output_comp.keys())
    total_diff = 0
    for el in elements:
        target_val = target_comp.get(el, 0)
        output_val = output_comp.get(el, 0)
        total_diff += abs(target_val - output_val)
    score = 1 - total_diff / 2  # Max total diff is 2
    return round(score, 3)

def calculate_chemical_factors_from_inputs(target_formula, interpretations):
    """
    Compute composition balance scores for interpretations using target formula and interpretation dictionary.
    
    Parameters:
        target_formula (str): The target compound formula (e.g., 'NiFe2O4').
        interpretations (dict): Dictionary with interpretations, each containing:
            - "phases": list of phase formulas
            - "weight_fraction": list of corresponding weight fractions
        
    Returns:
        dict: Updated interpretations with added 'balance_score'.
    """
    print("\nðŸŽ¯ Target Composition:", target_formula)
    elements_to_remove = ["O", "C", "N", "H"]

    # Normalize target composition
    target_composition = Composition(target_formula)
    filtered_target = {el: amt for el, amt in target_composition.get_el_amt_dict().items() if el not in elements_to_remove}
    normalized_target = Composition(filtered_target).fractional_composition
    print("ðŸ§ª Normalized Target Composition:", normalized_target)

    for interpretation_name, interpretation in interpretations.items():
        print(f"\nðŸ” Processing Interpretation: {interpretation_name}")
        phases = interpretation["phases"]
        weights = interpretation["weight_fraction"]

        combined_composition = {}
        for phase, weight in zip(phases, weights):
            try:
                comp = Composition(phase)
                filtered = {el: amt for el, amt in comp.get_el_amt_dict().items() if el not in elements_to_remove}
                norm_comp = Composition(filtered).fractional_composition
                for el, amt in norm_comp.items():
                    combined_composition[el] = combined_composition.get(el, 0) + amt * (weight / 100)
            except Exception as e:
                print(f"âŒ Error processing {phase}: {e}")

        total = sum(combined_composition.values())
        normalized_output = {el: round(amt / total, 3) for el, amt in combined_composition.items()}
        print("ðŸ“¦ Normalized Output Composition:", normalized_output)

        score = calculate_composition_balance_score_refined(normalized_target, normalized_output)
        interpretation["balance_score"] = score
        print("âœ… Balance Score:", score)

    return interpretations
```

```python
target = "NiFe2O4"

interpretations = {
    "Case 1": {
        "phases": ["NiFe2O4", "MgFeAlO4", "MgFe2O4", "Fe2SiO4"],
        "weight_fraction": [20, 2, 30, 48]
    },
    "Case 2": {
        "phases": ["NiFe2O4", "NiO"],
        "weight_fraction": [90, 10]
    },
    "Case 3": {
        "phases": ["NiFe2O4", "Fe2O3", "NiO"],
        "weight_fraction": [70, 20, 10]
    }
}

results = calculate_chemical_factors_from_inputs(target, interpretations)
```
