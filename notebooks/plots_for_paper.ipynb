```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
data = np.array([20, 27.5, 40, 27.5])

# Compute mean and standard deviation
mu = np.mean(data)  # Mean ()
sigma = np.std(data, ddof=1)  # Standard deviation (), using sample std dev (ddof=1)

# Define the normal PDF function
def normal_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Compute PDF values for the given data points
pdf_at_data_points = normal_pdf(data, mu, sigma)

# Display results
for x, pdf in zip(data, pdf_at_data_points):
    print(f"PDF at {x}: {pdf*100:.5f}")

# Plot the normal PDF
x_values = np.linspace(min(data) - 10, max(data) + 10, 100)
pdf_values = normal_pdf(x_values, mu, sigma)

plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, label="Estimated PDF (Normal Distribution)", color="blue")
plt.scatter(data, pdf_at_data_points, color="red", label="Data Points", zorder=3)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Probability Density Function (Normal Distribution)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_contributions(interpretations):
    """
    Compute the contributions of Statistical Factor, LLM Likelihood, and Composition Balance 
    to the final posterior probability.

    Parameters:
        interpretations (dict): Dictionary containing interpretation data.

    Returns:
        dict: Contributions of each factor to the final probability.
    """
    labels = []
    statistical_factors = []
    llm_likelihoods = []
    composition_balances = []
    
    for idx, (interpretation_name, interpretation) in enumerate(interpretations.items(), start=1):
        P_X = interpretation["X"] / 100  # Scale X to 0-1
        P_LLM = interpretation["LLM_interpretation_likelihood"]
        P_comp = interpretation["balance_score"]

        # Compute prior probability (Chemical Feasibility)
        P_prior = (P_LLM + P_comp) / 2

        # Compute final probability
        P_final = P_prior * P_X

        # Store values
        labels.append(f"Interpretation_{idx}")
        statistical_factors.append(P_X)
        llm_likelihoods.append(P_LLM)
        composition_balances.append(P_comp)

    # Convert to NumPy arrays for calculations
    statistical_factors = np.array(statistical_factors)
    llm_likelihoods = np.array(llm_likelihoods)
    composition_balances = np.array(composition_balances)

    # Normalize the values
    statistical_factors = statistical_factors / statistical_factors.sum()
    llm_likelihoods = llm_likelihoods / llm_likelihoods.sum()
    composition_balances = composition_balances / composition_balances.sum()

    # Weights for combining factors (set to 1 for equal contribution)
    w_s, w_l, w_c = 1, 1, 1

    # Compute final probability
    final_probabilities = (
        w_s * statistical_factors + w_l * llm_likelihoods + w_c * composition_balances
    )

    # Compute contributions as percentages
    contrib_statistical = (w_s * statistical_factors) / final_probabilities * 100
    contrib_llm = (w_l * llm_likelihoods) / final_probabilities * 100
    contrib_composition = (w_c * composition_balances) / final_probabilities * 100

    return labels, contrib_statistical, contrib_llm, contrib_composition


def plot_contributions(labels, contrib_statistical, contrib_llm, contrib_composition):
    """
    Plot the stacked bar chart showing the contributions of different factors.

    Parameters:
        labels (list): List of interpretation names.
        contrib_statistical (list): Contribution percentages of Statistical Factor.
        contrib_llm (list): Contribution percentages of LLM.
        contrib_composition (list): Contribution percentages of Composition Balance.
    """
    fig, ax = plt.subplots(figsize=(8, 6))
    indices = np.arange(len(labels))

    ax.barh(indices, contrib_statistical, label="Statistical Factor", color="lightblue")
    ax.barh(indices, contrib_llm, left=contrib_statistical, label="LLM Contribution", color="teal")
    ax.barh(indices, contrib_composition, left=contrib_statistical + contrib_llm, label="Composition Balance", color="mediumseagreen")

    # Annotate percentages on the bars
    for i in range(len(labels)):
        ax.text(contrib_statistical[i] / 2, i, f"{contrib_statistical[i]:.1f}%", va='center', ha='center', color="black", fontsize=10, weight='bold')
        ax.text(contrib_statistical[i] + contrib_llm[i] / 2, i, f"{contrib_llm[i]:.1f}%", va='center', ha='center', color="black", fontsize=10, weight='bold')
        ax.text(contrib_statistical[i] + contrib_llm[i] + contrib_composition[i] / 2, i, f"{contrib_composition[i]:.1f}%", va='center', ha='center', color="black", fontsize=10, weight='bold')

    ax.set_yticks(indices)
    ax.set_yticklabels(labels)
    ax.set_xlabel("Contribution to Final Probability (%)", fontsize=16)

    # Adjust tick parameters for axis
    ax.tick_params(axis='x', labelsize=14)  # X-axis tick labels size
    ax.tick_params(axis='y', labelsize=16)  # Y-axis tick labels size

    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Move legend to lower right with larger font size
    ax.legend(loc="lower right", bbox_to_anchor=(1.4, 0), fontsize=12)

    plt.show()
```

```python
import re
from collections import defaultdict

def parse_formula(formula, multiplier=1):
    """Recursively parse a chemical formula, applying multipliers to elements."""
    elements = defaultdict(int)
    i = 0
    while i < len(formula):
        if formula[i] == '(':
            depth = 1
            j = i + 1
            while j < len(formula) and depth > 0:
                if formula[j] == '(':
                    depth += 1
                elif formula[j] == ')':
                    depth -= 1
                j += 1
            inner_formula = formula[i+1:j-1]
            match = re.match(r'(\d+)', formula[j:])
            inner_multiplier = int(match.group(1)) if match else 1
            j += len(match.group(1)) if match else 0
            sub_elements = parse_formula(inner_formula, multiplier * inner_multiplier)
            for el, count in sub_elements.items():
                elements[el] += count
            i = j
        else:
            match = re.match(r'([A-Z][a-z]*)(\d*)', formula[i:])
            if not match:
                raise ValueError(f"Invalid formula at position {i}: {formula[i:]}")
            el = match.group(1).capitalize()
            count = int(match.group(2)) if match.group(2) else 1
            elements[el] += count * multiplier
            i += len(match.group(0))
    return elements


def flatten_formula(formula):
    parsed = parse_formula(formula)
    return ''.join(f'{el}{count}' for el, count in parsed.items())

# Example usage
phase = flatten_formula("Ti31(Fe21O52)3")
phase
```

```python
from pymatgen.core import Composition
from math import gcd
from functools import reduce

def get_empirical_formula(formula):
    comp = Composition(formula)
    el_amt_dict = comp.get_el_amt_dict()  # keys are strings like 'Mn'
    atom_counts = [round(val) for val in el_amt_dict.values()]
    
    # Compute GCD of all counts
    def compute_gcd(lst):
        return reduce(gcd, lst)
    
    factor = compute_gcd(atom_counts)
    
    # Normalize by GCD
    simplified = {el: round(count / factor, 6) for el, count in el_amt_dict.items()}
    return simplified
```

```python
import re
from collections import defaultdict, OrderedDict

def flatten_chemical_formula(formula):
    """
    Flatten a chemical formula by expanding parentheses and combining duplicate elements.
    Removes 1s, preserves element order.
    """
    def expand_parentheses(f):
        pattern = r'\(([^()]+)\)(\d*)'
        while re.search(pattern, f):
            f = re.sub(pattern, lambda m: expand_group(m.group(1), m.group(2)), f)
        return f

    def expand_group(group, multiplier):
        multiplier = int(multiplier) if multiplier else 1
        elems = re.findall(r'([A-Z][a-z]*)(\d*)', group)
        return ''.join(f"{el}{int(cnt or 1)*multiplier}" for el, cnt in elems)

    def parse_and_collapse(flat_formula):
        elems = re.findall(r'([A-Z][a-z]*)(\d*)', flat_formula)
        counts = OrderedDict()
        for el, cnt in elems:
            cnt = int(cnt or 1)
            counts[el] = counts.get(el, 0) + cnt
        return counts

    no_parens = expand_parentheses(formula)
    collapsed = parse_and_collapse(no_parens)
    return ''.join(f"{el}{cnt if cnt > 1 else ''}" for el, cnt in collapsed.items())
```

```python
from pymatgen.core import Composition

def describe_clean_composition(formula_str, digits=4, max_denominator=30):
    comp = Composition(formula_str)

    # Get rounded fractional composition
    frac_dict = {el: round(amt, digits) for el, amt in comp.fractional_composition.get_el_amt_dict().items()}

    # Reconstruct and quantize
    frac_comp = Composition(frac_dict)
    quantized_formula_str = frac_comp.get_integer_formula_and_factor(max_denominator=max_denominator)[0]
    quantized_dict = Composition(quantized_formula_str).get_el_amt_dict()

    # Format with clean chemical formula style
    ordered_elements = list(frac_dict.keys())
    quantized_str = ''.join(
        f"{el}" if int(quantized_dict[el]) == 1 else f"{el}{int(quantized_dict[el])}"
        for el in ordered_elements if el in quantized_dict
    )

    return f"{formula_str}, fractional_composition = {frac_dict}, approximately equal to {quantized_str}"
```

```python
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import numpy as np

# Parameters and weights
params = ["LLM likelihood", "Balance score", "Rwp", "Peak match score"]
weights = np.array([0.5, 0.7, 1.0, 0.5])


# Choose a colormap and normalize values to [0,1]
cmap = get_cmap("GnBu")  # try "plasma", "coolwarm", etc.

epsilon = 0.3  # ensures nothing maps below 0.1
norm_weights = (weights - weights.min()) / (weights.max() - weights.min())
norm_weights = norm_weights * (1 - epsilon) + epsilon  # scale into [0.1, 1]
colors = cmap(norm_weights)

# Create horizontal bar chart
plt.figure(figsize=(8, 5))
bars = plt.barh(params, weights, color=colors)
plt.title("Final optimized weights for AIF parameters", fontsize=18)
plt.xlabel("Weight Value", fontsize=16)
plt.xlim(0, 1)

# Add labels on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.03, bar.get_y() + bar.get_height()/2.0, f'{width:.1f}', va='center')
# Remove all spines (plot frame)
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)
# plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tick_params(axis='both', labelsize=14)
plt.tight_layout()
plt.show()
```
